```python

import numpy as np

def quickSort(arr, low, high):
    if low < high:
        # Выбираем pivot как средний элемент
        mid = (low + high) // 2
        # Слайс подмассива
        sub = arr[low:high + 1]
        # Относительный kth для partition
        kth_rel = mid - low
        # Partition (in-place, меняет arr)
        sub.partition(kth_rel)
        # Абсолютный pivot index
        pi = low + kth_rel
        # Рекурсия
        quickSort(arr, low, pi - 1)
        quickSort(arr, pi + 1, high)
    return arr  # Возвращаем отсортированный массив

# Тест
arr = np.array([10, 7, 8, 9, 1, 5, 6, 3])
print(quickSort(arr, 0, len(arr) - 1))  # [1 3 5 6 7 8 9 10]
```

```python
# Корректная версия: чтение одной температуры или спискаe
import numbers

# Ввод: можно ввести одно число (например, 23.5) или список чисел через запятую (например: 5, 12, 20)
raw = input("Введите температуру(ы) в градусах Цельсия (или список через запятую): ")

# Попробуем распарсить ввод: если есть запятая — делаем список, иначе одно число
if "," in raw:
    try:
        temperature = [float(x.strip()) for x in raw.split(",") if x.strip() != ""]
    except ValueError:
        raise ValueError("Неверный формат ввода: используйте числа или список чисел через запятую")
else:
    try:
        temperature = float(raw)
    except ValueError:
        raise ValueError("Неверный формат ввода: ожидается число")


def check_temperature(temperature):
    """Принимает single number или iterable чисел. Для некорректных значений печатает/возвращает сообщение об ошибке."""
    # Нормализуем: сделаем iterable
    if isinstance(temperature, numbers.Number):
        temps = [temperature]
    elif isinstance(temperature, Iterable):
        temps = list(temperature)
    else:
        raise TypeError("temperature должен быть числом или итерируемой последовательностью чисел")

    for t in temps:
        # валидация
        if not isinstance(t, numbers.Number):
            print("Пропускаем: значение не число:", t)
            continue
        if t < -273:
            print(f"Температура {t} ниже абсолютного нуля (-273). Пропускаем.")
            continue
        if t > 1000:
            print(f"Температура {t} слишком велика (>1000). Пропускаем.")
            continue

        # классификация
        if t < 10:
            print(f"{t}°C — Холодно")
        elif 10 <= t <= 25:
            print(f"{t}°C — Тепло")
        else:
            print(f"{t}°C — Жарко")


# Вызов
check_temperature(temperature)
```

```python
dice = range(1, 7)

outcomes = []
for a in dice:
    for b in dice:
        outcomes.append((a, b))


favorable = [ (a, b) for a, b in outcomes if a * b == 6 ]


total = len(outcomes)
fav_count = len(favorable)
probability = fav_count / total
rounded_probability = round(probability, 2)



print("Всего исходов:", total)
print("Благоприятных исходов:", fav_count)
print("Вероятность суммы 6 при броске двух кубиков:", rounded_probability)
print(f"Округленная вероятность: {rounded_probability}")


```
